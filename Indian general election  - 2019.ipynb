**1) Which political party won maximum constituency/seat in election?**

```python
A = df[['WINNER', 'PARTY']]
b = A.loc[A.WINNER == 1]
c = b['PARTY'].value_counts()
pd.DataFrame(c)
```

BJP(Bhartiya Janata Party) had won the maximum constituency followed by INC(Indian National congress) 

#### 2) which State has maximum constituency?

```python
r = df[['WINNER', 'STATE']]
t= r.loc[A.WINNER == 1]
b =t["STATE"].value_counts()
d = b.rename_axis('State').to_frame('seats')
pd.DataFrame(d)
```

Maximum Constituency are there in Uttar Pradesh with 78 constituency and followed by maharastra with 47 seats.

**3)How many male and female candidate won the election?**

```python
df_3 = df[['WINNER', 'GENDER']]
df_3.loc[A.WINNER == 1]
df.GENDER.value_counts()
```

Male have won the 463 seats and feamle won the 76 seats, which also show the lack of female representative.
