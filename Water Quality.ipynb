### Data Wrangling/ Data Cleansing

```python
ph = df.ph.mean()
```

```python
ph_mean = round(ph,4)
ph_mean
```

```python
df['ph'].fillna(value = ph_mean, inplace = True)       # replacing NaN value with mean value 
```

```python
Sulfate_mean = round(df['Sulfate'].mean(),4)
Sulfate_mean
```

```python
df['Sulfate'].fillna(Sulfate_mean, inplace = True)
```

```python
 Tri_mean = round(df['Trihalomethanes'].mean(), 4)
```

```python
df['Trihalomethanes'].fillna(Tri_mean, inplace = True)
```

```python
df.Potability= df.Potability.astype(float)
```

```python
df.info()
```

We converted the int dtype to float for Potability variable.
